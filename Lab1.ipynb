```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

fig, ax = plt.subplots(1, 1, figsize=(5, 5))

x = np.linspace(-10, 10, 1000)

ppf_one = stats.norm.ppf(x, loc=0, scale=1)
ppf_three = stats.norm.ppf(x, loc=0, scale=3)
ppf_half = stats.norm.ppf(x, loc=0, scale=0.5)

ax.set_title("Standard Normal Distribution PPF")
ax.plot(x, ppf_one, label='$\sigma$=1.0')
ax.plot(x, ppf_three, label='$\sigma$=3.0')
ax.plot(x, ppf_half, label='$\sigma$=0.5')
ax.set_xlim(0, 1)
ax.legend(loc='best')

plt.show()

one_sigma = 1 - stats.norm.cdf(1, loc=0, scale=1)
three_sigma = 1 - stats.norm.cdf(3, loc=0, scale=1)
five_sigma = 1 - stats.norm.cdf(5, loc=0, scale=1)

print("\u03C3 = 1 at 1\u03C3: " + str(stats.norm.ppf(one_sigma, loc=0, scale=1)))
print("\u03C3 = 1 at 3\u03C3: " + str(stats.norm.ppf(three_sigma, loc=0, scale=1)))
print("\u03C3 = 1 at 5\u03C3: " + str(stats.norm.ppf(five_sigma, loc=0, scale=1)))

half_sigma = 1 - stats.norm.cdf(0.5, loc=0, scale=0.6)
print("\u03C3 = 0.5 at 1\u03C3: " + str(stats.norm.ppf(half_sigma, loc=0, scale=0.6)))
```

__D.__ The negative side just means the point is on the left hand side of the PDF and is a feature of the integration, since the Standard Normal Distribution is symmetric about zero the sigmas will be the same on both sides of zero which is why the calculations above returned the sigma value given but negative.

### Question 3

__A.__ Hypothetical measurement is 12.5.

__B.__ What is the probability that my background will give me a value that looks like my signal  of 12.5 or greater?

__C.__ $\displaystyle \textrm{Probability} = \int_{12.5}^{\infty} \textrm{pdf} (x)\, dx$ or $\displaystyle \textrm{probability} = 1 - \left(\int_{0}^{12.5} \textrm{pdf} (x)\, dx\right)$ since the pdf is normalized to 1, or 100%, and is well defined from 0 to 12.5 but not from 12.5 to $\infty$.

__D.__

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

df = 5
d = stats.chi2.rvs(df, size=100000)

fig, ax = plt.subplots(1, 1)
x = np.linspace(0, 20, 10000)
cdf = stats.chi2.cdf(x, df)
ax.plot(x, cdf)
plt.xlim(0, 20)
print("Probability of signal at 12.5 not being from background = " + str(stats.chi2.cdf(12.5, df)))
print("ie. probability =  " + str(1 - stats.chi2.cdf(12.5, df)) + "that the background looks like the 12.5 signal.")
print(str(stats.chi2.ppf(1 - stats.chi2.cdf(12.5, df), df)) + " \u03C3")
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

x = np.linspace(0, 20, 10000)
pdf_one = stats.chi2.pdf(12.5, 1)
pdf_five0 = stats.chi2.pdf(12.5, 5)
pdf_ten = stats.chi2.pdf(12.5, 10)
print("df = 1: Probability at 12.5 = " + str(stats.chi2.ppf(pdf_one, 1)))
print("df = 5: Probability at 12.5 = " + str(stats.chi2.ppf(pdf_five0, 5)))
print("df = 10: Probability at 12.5 = " + str(stats.chi2.ppf(pdf_ten, 10)))
print()
pdf_five1 = stats.chi2.pdf(5, 5)
pdf_five2 = stats.chi2.pdf(10, 5)
pdf_five3 = stats.chi2.pdf(15, 5)
print("df = 5: Probability at 5 = " + str(stats.chi2.ppf(pdf_five1, 5)))
print("df = 5: Probability at 10 = " + str(stats.chi2.ppf(pdf_five2, 5)))
print("df = 5: Probability at 15 = " + str(stats.chi2.ppf(pdf_five3, 5)))
```

As the df increses the PDF gets stretched out, as seen in graphs above, so 12.5 becomes more and more probable, but would eventually become less probable as df continues to increase. For a distribution of df=5 the most likely values are around 2.5 to 5 and as the points get larger they become less and less likely.

## Non-Continuous Distribution 

### Question 1

__A.__ Poisson Distribution

The Poisson distribution is often used to describe the number of times an event occured in a given interval. The Poisson distribution takes in one parameter, $\mu\$, the shape parameter which is the average rate of events per unit interval.

__B.__

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

x = np.arange(0, 100)

pmf0 = stats.poisson.pmf(x, 10)
pmf1 = stats.poisson.pmf(x, 20)
pmf2 = stats.poisson.pmf(x, 50)
pmf3 = stats.poisson.pmf(x, 80)

plt.step(x, pmf0, label='mu=10')
plt.step(x, pmf1, label='mu=20')
plt.step(x, pmf2, label='mu=50')
plt.step(x, pmf3, label='mu=80')
plt.legend(loc='best')
plt.xlim(0, 100)
plt.ylim(0, 0.15)

plt.show()
```

__C.__ In Seattle is rains on average 152 days a year. This can be modeled by a Poisson distribution with a shape parameter of 152. What is the probability of it only raining 120 days in a year in Seattle?

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats

x = np.arange(0, 200)

pmf1 = stats.poisson.pmf(x, 152)

plt.step(x, pmf1, label='$\mu$=152')

plt.legend(loc='best')
plt.xlim(0, 200)
plt.ylim(0, 0.05)

plt.show()

probability = stats.chi2.cdf(120, 152)
print("Probability = " + str(probability))
```

__D.__ The x-axis on the pdf graph would represent the potential values one could get in their experiment which by definition of the Poisson distribution, have to be descrete. In our example above we are looking at the probability of a rainy day and for our purposes it's either raining or its not raining, we don't care about a 0.25 rainy day, which makes our experiment descrete. This also means that our sigma has be be discrete since our accuracy can't be within half of a rainy day since we don't measure to a half of a rainy day.

__E.__ Even though the results of the Poisson distribution are descrete you can still have a mean of 9.2, per the example in the lab question. This means that the mean falls inbetween 9 and 10, being more probable to get 9, since 9.2 is closer to 9 than 10.
